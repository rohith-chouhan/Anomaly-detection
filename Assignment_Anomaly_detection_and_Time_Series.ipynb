### 1) What is Dimensionality Reduction? Why is it important in machine learning?
#### > Dimensionality Reduction is a technique used in machine learning to reduce the number of input features (dimensions) in a dataset while preserving as much important information as possible.It means transforming high-dimensional data (many features) into a lower-dimensional space (fewer features).
#### > Dimensionality Reduction Important in Machine Learning because :
##### **Reduces Overfitting** : Fewer features → less noise → model generalizes better.
##### **Improves Model Performance** : High dimensionality can make training slow and inefficient (curse of dimensionality).
##### **Speeds Up Training** : Less data → faster computation.
##### **Helps Visualization** : Data with
- 100+ dimensions is impossible to visualize
- PCA or t-SNE reduces it to 2D or 3D for plotting.
##### **Removes Redundancy** : Many features may be
- Highly correlated
- Irrelevant
- Repetitive
- Dimensionality reduction compresses them.
##### **Improves Storage Efficiency** : Smaller feature space → less memory needed.

### 2) Name and briefly describe three common dimensionality reduction techniques.
#### **Principal Component Analysis (PCA)** : PCA is a linear dimensionality reduction technique that transforms the original features into a new set of features called principal components.These components capture the maximum variance in the data.
- Converts correlated features into uncorrelated ones
- Keeps only the components that explain most of the variance
- Commonly used for visualization and noise reduction

#### **t-Distributed Stochastic Neighbor Embedding (t-SNE)** : t-SNE is a non-linear technique mainly used for visualization in 2D or 3D.
- Preserves local structure (nearby points stay close)
- Excellent for visualizing clusters
- Often used for high-dimensional data like images or word embeddings
- Not ideal for large datasets (slow)
#### **Linear Discriminant Analysis (LDA)** : LDA is a supervised dimensionality reduction technique.
- Considers class labels to maximize class separability
- Projects data onto a lower-dimensional space where classes are spread apart
- Often used before classification tasks (e.g., face recognition)

### 3) What is clustering in unsupervised learning? Mention three popular clustering algorithms.
#### > Clustering in unsupervised learning is the process of grouping similar data points together based on their patterns or features—without using labeled data.Clustering helps discover hidden patterns, natural groupings, and structure in data.The goal is to ensure:
- Points within the same cluster are similar
- Points in different clusters are dissimilar
#### Three Popular Clustering Algorithms
##### **K-Means Clustering** :
- Divides data into K clusters based on distance to cluster centroids.
- Fast and widely used.
- Works well for spherical/compact clusters.
##### **Hierarchical Clustering** :
- Builds clusters in a tree-like structure (dendrogram).
##### Two types:
- Agglomerative (bottom-up)
- Divisive (top-down)
##### **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** :
- Groups points based on density.
- Detects clusters of complex shapes.
- Identifies noise/outliers.
- Good for uneven cluster sizes.


### 4) Explain the concept of anomaly detection and its significance.
#### > Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the norm.These unusual observations are called anomalies or outliers.Anomaly detection is typically used in unsupervised learning, because anomalies are rare and labels are often unavailable.It involves finding rare or unexpected items in data that:
- Do not follow the general pattern
- May indicate critical events
- Are often very few compared to normal data
#### Significance of Anomaly Detection
#### **Fraud Detection** :
- Detecting unusual credit card transactions
- Spotting fake insurance claims
##### **Network and Cybersecurity** :
- Identifying unusual login patterns
- Detecting intrusion or malware
##### **Fault Detection in Machines** :
- Abnormal temperature, vibration, or pressure in equipment
- Helps in predictive maintenance
##### **Medical and Health Monitoring** :
- Detecting abnormal heartbeat or lab test values
##### **Quality Control in Manufacturing** :
- Identifying defective products in production lines
##### **Data Cleaning** :
- Removing outliers that can harm model accuracy



### 5) List and briefly describe three types of anomaly detection techniques.
#### **Statistical (Rule-Based) Methods** : These methods assume that normal data follows a certain statistical distribution (example: Gaussian).Any data point that falls too far from the expected range is flagged as an anomaly.
##### **Examples** :
- Z-score
- Gaussian distribution modeling
- Interquartile range (IQR)
##### Use case: Simple datasets with predictable patterns.

#### **Distance-Based Methods** : These methods detect anomalies based on how far a point is from others.Anomalies usually have large distances from their nearest neighbors.
##### **Examples** :
- k-Nearest Neighbors (kNN) anomaly detection
- Local Outlier Factor (LOF)
##### Use case: When normal points form dense clusters.
#### **Machine Learning–Based Methods** : These methods learn a model of “normal behavior” and detect points that do not fit this behavior.
##### **Examples** :
- Isolation Forest (isolates anomalies quickly)
- One-Class SVM (learns a boundary around normal data)
- Autoencoders (deep learning reconstruction error)
##### Use case: Complex, high-dimensional data (images, logs, transactions).

### 6) What is time series analysis? Mention two key components of time series data.
#### > Time Series Analysis is a method of analyzing data points collected or recorded over time at regular intervals.The main goal is to understand patterns, trends, and seasonal behavior to make forecasts or insights about future values.Time series analysis helps identify how values evolve over time and predict future movements.
#### Two Key Components of Time Series Data
##### **Trend** : A long-term increase or decrease in the data over time.
##### Example : Company revenue steadily rising over several years.
##### **Seasonality** : Regular, repeating patterns or cycles in the data.
##### Example : Higher ice cream sales every summer or increased shopping during festivals.

### 7) Describe the difference between seasonality and cyclic behavior in time series.
#### **Seasonality** : Seasonality refers to regular, repeating patterns in a time series that occur at fixed and known intervals.
#### **Characteristics** :
- Predictable and happens at the same time each period
- Fixed frequency (daily, monthly, yearly, etc.)
- Driven by calendar or seasonal factors
#### **Examples** :
- Higher ice-cream sales every summer

- Increased electricity usage every evening

- Online shopping spikes every December
#### **Cyclic Behavior** : Cyclic behavior refers to patterns that occur in long-term cycles, but not at fixed intervals.
#### **Characteristics** :
- Duration and intensity vary
- Not based on the calendar
- Influenced by economic or external factors
- Harder to predict
#### **Examples** :
- Economic booms and recessions
- Real estate market cycles
- Long-term business growth and decline cycles

### 8) Write Python code to perform K-means clustering on a sample dataset.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

labels = kmeans.labels_

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("K-Means Clustering")
plt.show()
```

### 9) What is inheritance in OOP? Provide a simple example in Python.
#### > Inheritance in Object-Oriented Programming (OOP) is a feature that allows one class (child or derived class) to inherit attributes and methods from another class (parent or base class).It helps in code reusability, organization, and reducing duplication.
### Example is provided below :

```python
class Animal:
    def speak(self):
        return "This animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "The dog barks"

a = Animal()
d = Dog()

print(a.speak())
print(d.speak())
```

### 10) How can time series analysis be used for anomaly detection.
#### > Time series analysis can be used for anomaly detection by identifying data points that do not follow the normal time-based pattern of the series.
#### **How it works** :
- Model the normal behavior of the time series
(Using moving averages, ARIMA, forecasting models, etc.)
- Compare actual values with predicted/expected values
- Large deviations from the expected pattern are flagged as anomalies.
#### **Examples** :
- A sudden spike in website traffic
- An unexpected drop in sales
- Abnormal temperature in a machine sensor
